```python
from statsmodels.stats.proportion import proportions_ztest

control_clicks = 154303
control_purchases = 15161

test_clicks = 180970
test_purchases = 15637

count = [control_purchases, test_purchases]
nobs = [control_clicks, test_clicks]

stat, pval = proportions_ztest(count, nobs)

print("Z-statistic:", stat)
print("P-value:", pval)
```

```python
import numpy as np

# Data
control_success = 15161
control_total = 154303

test_success = 15637
test_total = 180970

# Posterior sampling (Beta distribution)
control_samples = np.random.beta(control_success + 1, control_total - control_success + 1, 100000)
test_samples = np.random.beta(test_success + 1, test_total - test_success + 1, 100000)

# Probability test is better
prob_test_better = np.mean(test_samples > control_samples)

print("Probability Test is Better:", prob_test_better)
```
